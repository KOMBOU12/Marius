 fonction x(t)  = sin(6pi t)

```python
import numpy as np
import matplotlib.pyplot as plt

# Définir la fonction x(t) = sin(6πt)
t = np.linspace(0, 1, 500)
# Initialisation des paramètres
N = 100 # Nombre d'echantillons
grille = np.linspace(0, 1, N)  # Grille entre 0 et 1
sigma_matrix = np.random.normal(0, 0.2, (N, len(t)))  # Bruit blanc pour chaque réalisation
```

```python
x = np.sin(6 * np.pi * grille)
```

```python
plt.plot(grille, x)
plt.show()
```

On simule $y(t_i) = x(t_i) + ϵ_i$

```python
y = x +  np.random.normal(loc=0, scale=0.25, size=N)
```

```python
plt.scatter(grille, y, color='red', alpha=0.4, label='Star')
plt.plot(grille, x, color='blue', label='Line')
plt.legend()

plt.show()
```

Ici on détermine la matrice $\phi$ qui donne $t_i^0 , t_i^1, t_i^2, ...t_i^8$

```python
# import numpy as np
# phi = np.zeros((N, 8))  # Initialisation de la matrice phi avec des zéros
```

```python
# for i in range(N):
#     phi[i] = np.array([t[i]**j for j in range(8)])
```

```python
K = 8
# Calcul de la matrice Phi
Phi = np.array([[t**k for k in range(1, K+1)] for t in grille])
```

```python
plt.plot(grille, Phi)
plt.title("Matplot de Phi en Python")
plt.show()
```

```python
# Calcul de c_hat
c_hat = np.linalg.inv(Phi.T @ Phi) @ Phi.T @ y

# Définir la fonction y_hat
def y_hat(t):
    Phi_t = np.array([[t_i**k for k in range(1, K+1)] for t_i in t])
    return Phi_t @ c_hat

# Calculer y_hat pour les points de la grille
y_hat_values = y_hat(grille)

# Afficher les résultats
print("Coefficients c_hat:", c_hat)
```
