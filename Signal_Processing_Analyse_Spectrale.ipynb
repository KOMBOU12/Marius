```python
import numpy as np
import matplotlib.pyplot as plt

# simulation d'une suite périodique
y1 = np.tile([1, 0, 0], 80)          # repeter motif 80 fois
y2 = np.tile([0, 0, 0, 2], 60)       # repeter 60 fois
y3 = np.tile([0, 0, 0, 0, 0, 5], 40) # repeter 40 fois

y = y1 + y2 + y3
print(y[0:20])


# Tracé du graphique
plt.figure(figsize=(10, 6))
plt.stem(range(len(y)), y, basefmt=" ")
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Sum of Sequences")
plt.show()

plt.plot(y[0:36])
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Calcul de la transformée de Fourier et du spectre de puissance
I = np.abs(np.fft.fft(y))**2 / 240  # FFT et puissance normalisée
P = (4 / 240) * I[:120]  # Normalisation de la puissance
f = np.arange(120) / 240  # Fréquences correspondantes

# Tracé du spectre de puissance
#plt.figure(figsize=(10, 6))
plt.plot(f[1:], P[1:], label="Spectre de puissance")  # Équivalent de type="l" en R
plt.xlabel("Frequency")
plt.ylabel("Power")
plt.title("Spectral Power")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Lecture des données à partir de l'URL
url = "https://web.stanford.edu/class/earthsys214/data/nycmeas.dat.txt"
meas = pd.read_table(url, header=None, sep = " ", names=["Date", "Cases"])

# Tracé de la série temporelle
plt.figure(figsize=(10, 6))
plt.plot(meas["Date"], meas["Cases"], label="Measles Cases", color="blue")
plt.xlabel("Date")
plt.ylabel("Cases")
plt.title("Measles Cases in NYC")
plt.legend()
plt.show()
```
