## **Distribution Normale Centrée (0, 0) Avec Écart-Type = 4**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Définir les paramètres de la distribution normale
mean = [0, 0]  # Moyenne (centre)
covariance = [[4**2, 0], [0, 4**2]]  # Matrice de covariance (écart-type 4 dans chaque direction)

# Créer une grille pour représenter la distribution
x, y = np.linspace(-20, 20, 500), np.linspace(-20, 20, 500)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))

# Calculer la densité de probabilité
rv = multivariate_normal(mean, covariance)
Z = rv.pdf(pos)

# Représenter la distribution normale
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, levels=30, cmap='viridis')
plt.colorbar(label="Densité de probabilité")
plt.title("Distribution normale centrée (0, 0) avec écart-type = 4")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis('equal')
plt.show()
```


## **Distribution Normale Centrée Sur Un Cercle de centre (0,0) et de rayon 4, avec un écart-type de 4.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres
radius = 4  # Rayon du cercle
std_dev = 4  # Écart-type
num_circle_points = 100  # Nombre de points le long du cercle
samples_per_point = 100  # Nombre d'échantillons par point du cercle

# Générer des angles pour les points sur le cercle
angles = np.linspace(0, 2 * np.pi, num_circle_points, endpoint=False)

# Liste pour stocker tous les échantillons
all_samples = []

for theta in angles:
    # Coordonnées du point sur le cercle
    circle_point = np.array([radius * np.cos(theta), radius * np.sin(theta)])

    # Générer des échantillons à partir d'une distribution normale centrée sur le point du cercle
    samples = np.random.normal(loc=circle_point, scale=std_dev, size=(samples_per_point, 2))

    all_samples.append(samples)

# Combiner tous les échantillons
all_samples = np.vstack(all_samples)

# Représenter la distribution
plt.figure(figsize=(8, 8))
plt.hist2d(all_samples[:, 0], all_samples[:, 1], bins=100, cmap='viridis')
plt.colorbar(label='Densité')
plt.title('Distribution normale autour d\'un cercle de rayon 4 avec écart-type 4')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis('equal')
plt.show()
```

```python
def draw_2d_simplicial_complex(simplices, pos=None, return_pos=False, ax = None):
    """
    Draw a simplicial complex up to dimension 2 from a list of simplices, as in [1].

        Args
        ----
        simplices: list of lists of integers
            List of simplices to draw. Sub-simplices are not needed (only maximal).
            For example, the 2-simplex [1,2,3] will automatically generate the three
            1-simplices [1,2],[2,3],[1,3] and the three 0-simplices [1],[2],[3].
            When a higher order simplex is entered only its sub-simplices
            up to D=2 will be drawn.

        pos: dict (default=None)
            If passed, this dictionary of positions d:(x,y) is used for placing the 0-simplices.
            The standard nx spring layour is used otherwise.

        ax: matplotlib.pyplot.axes (default=None)

        return_pos: dict (default=False)
            If True returns the dictionary of positions for the 0-simplices.

        References
        ----------
        .. [1] I. Iacopini, G. Petri, A. Barrat & V. Latora (2019)
               "Simplicial Models of Social Contagion".
               Nature communications, 10(1), 2485.
    """


    #List of 0-simplices
    nodes =list(set(itertools.chain(*simplices)))

    #List of 1-simplices
    edges = list(set(itertools.chain(*[[tuple(sorted((i, j))) for i, j in itertools.combinations(simplex,
2)] for simplex in simplices])))

    #List of 2-simplices
    triangles = list(set(itertools.chain(*[[tuple(sorted((i, j, k))) for i, j, k in itertools.combinations(simplex,
3)] for simplex in simplices])))

    if ax is None: ax = plt.gca()
    ax.set_xlim([-1.1, 1.1])
    ax.set_ylim([-1.1, 1.1])
    ax.get_xaxis().set_ticks([])
    ax.get_yaxis().set_ticks([])
    ax.axis('off')

    if pos is None:
        # Creating a networkx Graph from the edgelist
        G = nx.Graph()
        G.add_edges_from(edges)
        # Creating a dictionary for the position of the nodes
        pos = nx.spring_layout(G)

    # Drawing the edges
    for i, j in edges:
        (x0, y0) = pos[i]
        (x1, y1) = pos[j]
        line = plt.Line2D([ x0, x1 ], [y0, y1 ],color = 'black', zorder = 1, lw=0.7)
        ax.add_line(line);

    # Filling in the triangles
    for i, j, k in triangles:
        (x0, y0) = pos[i]
        (x1, y1) = pos[j]
        (x2, y2) = pos[k]
        tri = plt.Polygon([ [ x0, y0 ], [ x1, y1 ], [ x2, y2 ] ],
                          edgecolor = 'black', facecolor = plt.cm.Blues(0.6),
                          zorder = 2, alpha=0.4, lw=0.5)
        ax.add_patch(tri);

    # Drawing the nodes
    for i in nodes:
        (x, y) = pos[i]
        circ = plt.Circle([ x, y ], radius = 0.02, zorder = 3, lw=0.5,
                          edgecolor = 'Black', facecolor = u'#ff7f0e')
        ax.add_patch(circ);

    if return_pos: return pos
```
