## **Distribution Normale Centrée (0, 0) Avec Écart-Type = 4**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Définir les paramètres de la distribution normale
mean = [0, 0]  # Moyenne (centre)
covariance = [[4**2, 0], [0, 4**2]]  # Matrice de covariance (écart-type 4 dans chaque direction)

# Créer une grille pour représenter la distribution
x, y = np.linspace(-20, 20, 500), np.linspace(-20, 20, 500)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))

# Calculer la densité de probabilité
rv = multivariate_normal(mean, covariance)
Z = rv.pdf(pos)

# Représenter la distribution normale
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, levels=30, cmap='viridis')
plt.colorbar(label="Densité de probabilité")
plt.title("Distribution normale centrée (0, 0) avec écart-type = 4")
plt.xlabel("X")
plt.ylabel("Y")
plt.axis('equal')
plt.show()
```


## **Distribution Normale Centrée Sur Un Cercle de centre (0,0) et de rayon 4, avec un écart-type de 4.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres
radius = 4  # Rayon du cercle
std_dev = 4  # Écart-type
num_circle_points = 100  # Nombre de points le long du cercle
samples_per_point = 100  # Nombre d'échantillons par point du cercle

# Générer des angles pour les points sur le cercle
angles = np.linspace(0, 2 * np.pi, num_circle_points, endpoint=False)

# Liste pour stocker tous les échantillons
all_samples = []

for theta in angles:
    # Coordonnées du point sur le cercle
    circle_point = np.array([radius * np.cos(theta), radius * np.sin(theta)])

    # Générer des échantillons à partir d'une distribution normale centrée sur le point du cercle
    samples = np.random.normal(loc=circle_point, scale=std_dev, size=(samples_per_point, 2))

    all_samples.append(samples)

# Combiner tous les échantillons
all_samples = np.vstack(all_samples)

# Représenter la distribution
plt.figure(figsize=(8, 8))
plt.hist2d(all_samples[:, 0], all_samples[:, 1], bins=100, cmap='viridis')
plt.colorbar(label='Densité')
plt.title('Distribution normale autour d\'un cercle de rayon 4 avec écart-type 4')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis('equal')
plt.show()
```
